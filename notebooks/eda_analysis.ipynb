```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/raw/201306-citibike-tripdata.csv")
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['starttime'] = pd.to_datetime(df['starttime'])
df['stoptime'] = pd.to_datetime(df['stoptime'])
```

```python
pd.set_option('display.max_rows', None)
df['start station name'].value_counts()
```

```python
df['usertype'].value_counts().plot(kind='bar')
plt.title("User Types")
plt.show()
```

```python
# create hourly interval
df['start_hour'] = df['starttime'].dt.hour

# group and count
trips_per_start_station_hour = (
    df.groupby(['start station name', 'start_hour'])
      .size()
      .reset_index(name='trip_count')
      .sort_values('trip_count', ascending=False)
)
trips_per_start_station_hour
```
